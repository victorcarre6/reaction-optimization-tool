```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.rdFingerprintGenerator import GetMorganGenerator
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

df = pd.read_csv("datas/buchwald_full.csv")
```

```python
print("Aperçu des premières lignes :")
display(df.head())
```

```python
print("\nRésumé d'information :")
df.info()

print("\nValeurs manquantes :")
print(df.isnull().sum())
```

```python
print("\nStatistiques sur température et rendement :")
print(df[["temperature_Celsius", "yield_percent"]].describe())
```

```python
sns.histplot(df["temperature_Celsius"], bins=30, kde=True)
plt.title("Distribution des températures")
plt.xlabel("Température (°C)")
plt.ylabel("Fréquence")
plt.show()

sns.histplot(df["yield_percent"], bins=30, kde=True)
plt.title("Distribution des rendements")
plt.xlabel("Rendement (%)")
plt.ylabel("Fréquence")
plt.show()

sns.histplot(df["inputs_base_SMILES"], bins=30, kde=True)
plt.title("Distribution des bases")
plt.xlabel("Bases (Code SMILES")
plt.xticks(rotation=45, ha='right', va='top')
plt.ylabel("Fréquence")
plt.show()

sns.scatterplot(data=df, x="temperature_Celsius", y="yield_percent")
plt.title("Température vs Rendement")
plt.xlabel("Température (°C)")
plt.ylabel("Rendement (%)")
plt.show()
```

```python
# Visualisation impact de la base

plt.figure(figsize=(12,6))
sns.boxplot(x="yield_percent", y="inputs_base_SMILES", data=df)
plt.xticks(rotation=45)
plt.title("Distribution du rendement selon la base")
plt.show()

sns.scatterplot(x="yield_percent", y="inputs_base_SMILES", data=df)
plt.title("Distribution du rendement selon la base")
plt.xlabel("Rendement (%)")
plt.ylabel("Base (SMILES)")
plt.show()

plt.figure(figsize=(12,6))
sns.scatterplot(x="yield_percent", y="temperature_Celsius", hue="inputs_base_SMILES", data=df)
plt.title("Rendement en fonction de la température pour chaque base")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout()
plt.show()
```

```python
# Essai de vectorisation de la base (infructueux)

morgan_gen = GetMorganGenerator(radius=2, fpSize=2048)

df["mol_base"] = df["inputs_base_SMILES"].apply(lambda x: Chem.MolFromSmiles(x) if pd.notnull(x) else None)

def mol_to_fp(mol):
    if mol is None:
        return np.zeros((2048,))
    return np.array(morgan_gen.GetFingerprint(mol))

# Créer les vecteurs fingerprints
fp_array = np.array([mol_to_fp(mol) for mol in df["mol_base"]])
fp_df = pd.DataFrame(fp_array, columns=[f"fp_{i}" for i in range(fp_array.shape[1])])

# Simplification des vecteurs
pca = PCA(n_components=2)
base_pca = pca.fit_transform(fp_array)  # fp_array = fingerprints de la base

df["base_pca1"] = base_pca[:, 0]
df["base_pca2"] = base_pca[:, 1]

# Visualisation
plt.figure(figsize=(8,6))
sc = plt.scatter(df["temperature_Celsius"], df["base_pca1"], c=df["yield_percent"], cmap="viridis", s=50)
plt.colorbar(sc, label="Yield (%)")
plt.xlabel("Température (°C)")
plt.ylabel("Base (PCA 1)")
plt.title("Rendement en fonction de la base et de la température")
plt.show()


fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111, projection='3d')

scat = ax.scatter(
    df["temperature_Celsius"],
    df["base_pca1"],
    df["yield_percent"],
    c=df["yield_percent"],
    cmap="plasma",
    s=50,
    alpha=0.7
)

fig.colorbar(scat, label="Yield (%)")

ax.set_xlabel("Température (°C)")
ax.set_ylabel("Base PCA 1")
ax.set_zlabel("Rendement (%)")
plt.title("Température / Base (PCA1) / Rendement")

# Affichage des SMILES en annotations (ex: 20 premiers points)
for i, row in df.head(20).iterrows():
    ax.text(
        row["temperature_Celsius"],
        row["base_pca1"],
        row["yield_percent"],
        row["inputs_base_SMILES"],
        fontsize=8,
        rotation=20,
        color="black"
    )

plt.show()
```
